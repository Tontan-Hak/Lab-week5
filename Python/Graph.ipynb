# Line graph plote using Python

## Basic Line Graph using Matplotlib

```python
import matplotlib.pyplot as plt 
# Sample data
x = [1,2,3,4,5]
y = [2,3,5,7,11]
# Plotting the line graph 
plt.plot(x,y, marker = 'o')
plt.title('Basic Line Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

## Multiple Line Graph using Matplotlib

```python
import matplotlib.pyplot as plt 
# Sample data
x = [1,2,3,4,5]
y1 = [2,3,5,7,11]
y2 = [1,4,6,8,10]
# Plotting multiple line graph 
plt.plot(x,y1,label ='Line 1', marker = 'o')
plt.plot(x,y2, label='Line 2', marker='s')
plt.title('Basic Line Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

## Line Graph with Error Bars using Matplotlib

```python
import matplotlib.pyplot as plt 
import numpy as np
# Sample data
x = np.array([1,2,3,4,5])
y = np.array([2,3,5,7,11])
yerr = np.array([0.5,0.4,0.3,0.2,0.1])
# Plotting the line graph with error bars
plt.errorbar(x,y,yerr=yerr, fmt ='-o', capsize=5)
plt.title('Line Graph With Error Bars')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

## Line Graph using Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt 
# Sample data
data = {
'x' : [1,2,3,4,5],
'y' : [2,3,5,7,11]
}
# Create a Seaborn line plot
sns.lineplot(x='x',y='y', data =data, marker = 'o')
plt.title('Line Graph using Seaborn')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

## Interactive Line Grph using Plotly

```python
import plotly.express as px
# Sample data
data = {
'x' : [1,2,3,4,5],
'y' : [2,3,5,7,11]
}
# Create an interactive line plot
fig = px.line(data, x='x', y='y', title='Interactive Line Graph using plotly', markers = True)
fig.show()
```
