# Circle graph using Python

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the circle parameters
r = 3  # radius
x0 = 1  # x-center
y0 = 1  # y-center

# Generate some data
t = np.linspace(0, 2 * np.pi, 1000)
x = x0 + r * np.cos(t)
y = y0 + r * np.sin(t)

# Create a circle graph
fig, ax = plt.subplots()
ax.plot(x, y)

# Add labels and a title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Circle Graph')

# Show the plot
plt.show()
```

# Ellipse graph using Python

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the ellipse parameters
a = 3  # semi-major axis
b = 2  # semi-minor axis
x0 = 1  # x-center
y0 = 1  # y-center

# Generate some data
t = np.linspace(0, 2 * np.pi, 1000)
x = x0 + a * np.cos(t)
y = y0 + b * np.sin(t)

# Create an ellipse graph
fig, ax = plt.subplots()
ax.plot(x, y)

# Add labels and a title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Ellipse Graph')

# Show the plot
plt.show()
```

# Greatest integer function using Python

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the greatest integer function
def greatest_integer(x):
    return np.floor(x)

# Generate some data
x = np.linspace(-5, 5, 1000)
y = greatest_integer(x)

# Create a greatest integer graph
fig, ax = plt.subplots()
ax.plot(x, y)

# Add labels and a title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Greatest Integer Graph')

# Show the plot
plt.show()
```

# Quadratic Function using Python

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the quadratic function
def quadratic(x, a, b, c):
    return a*x**2 + b*x + c

# Generate some data
x = np.linspace(-10, 10, 1000)
y = quadratic(x, 4,2,1)

# Create a quadratic graph
fig, ax = plt.subplots()
ax.plot(x, y)

# Add labels and a title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Quadtatic Graph')

# Show the plot
plt.show()
```

# Hyperbola graph using Python

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some data with positive x-values only
x = np.linspace(-20, 20, 500)  # Avoid zero or negative numbers
y1 = np.sqrt(x**2 - 1)
y2 = -np.sqrt(x**2 - 1)

# Create a logarithm function graph
fig, ax = plt.subplots()
ax.plot(x, y1, color='blue', label = 'y = +sqrt(x^2 - 1)')
ax.plot(x, y2, color='red', label = 'y = -sqrt(x^2 - 1)')

# Add labels and a title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Hyperbola Graph')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data and mask invalid values
x = np.linspace(-20, 20, 500)
y1 = np.sqrt(x**2 - 1)
y2 = -np.sqrt(x**2 - 1)

# Mask the invalid regions where x**2 - 1 < 0
y1 = np.where(x**2 - 1 >= 0, y1, np.nan)
y2 = np.where(x**2 - 1 >= 0, y2, np.nan)

# Create the hyperbola graph
fig, ax = plt.subplots()
ax.plot(x, y1, color='blue', label='y = +sqrt(x^2 - 1)')
ax.plot(x, y2, color='red', label='y = -sqrt(x^2 - 1)')

# Add labels, title, and legend
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Hyperbola Graph')
ax.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x-values while ensuring |x| > 1 to avoid invalid values in sqrt
x = np.linspace(-20, 20, 500)
x_valid = x[np.abs(x) > 1]  # Filter out values where |x| <= 1

# Calculate y-values for the hyperbola
y1 = np.sqrt(x_valid**2 - 1)
y2 = -np.sqrt(x_valid**2 - 1)

# Create the hyperbola graph
fig, ax = plt.subplots()
ax.plot(x_valid, y1, color='blue', label='y = +sqrt(x^2 - 1)')
ax.plot(x_valid, y2, color='red', label='y = -sqrt(x^2 - 1)')

# Add labels, title, and legend
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Hyperbola Graph')
ax.legend()

# Show the plot
plt.show()
```

# Signum graph using Python

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the signum function for an array of x values
x = np.linspace(-10, 10, 1000)
y = np.sign(x)

# Create the graph
fig, ax = plt.subplots()
ax.plot(x, y, label='Signum Function')

# Add a horizontal line at y=0 for clarity
ax.axhline(0, color='black',linewidth=0.5)

# Add labels and title
ax.set_xlabel('x')
ax.set_ylabel('sign(x)')
ax.set_title('Signum Function Graph')

# Add grid for better visualization
ax.grid(True)

# Show the legend
ax.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the signum Function
def signum(x):
    if x > 0:
        return 1
    elif x == 0:
        return 0
    else:
        return -1

# Generate some data
x = np.linspace(-5, 5, 1000)
y = [signum(i) for i in x]

# Create a signum graph
fig, ax = plt.subplots()
ax.plot(x, y)

# Add labels and a title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Signum Graph')

# Show the plot
plt.show()
```
